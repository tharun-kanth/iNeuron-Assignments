#### 2. Why do raw strings often appear in Regex objects?

Ans - Regular expressions use the backslash character ('\') to indicate special forms (Metacharacters) or to allow special characters (speical sequences) to be used without invoking their special meaning. In other words, Raw strings are used so that backslashes do not have to be escaped.

#### 6. In standard expression syntax, parentheses and intervals have distinct meanings. How can you tell a regex that you want it to fit real parentheses and periods?

```python
import re
Sample = re.compile(r'(\(\d\d\d\)) (\d\d\d-\d\d\d\d)')
mo = Sample.search('My telephone number is (213) 444-3545.')
print(mo.group())
```

#### 12. What do you mean by the \d, \w, and \s shorthand character classes signify in regular expressions?

Ans -  The \d, \w, and \s shorthand character classes match a single digit, word, or space characters respectively.

\w – Matches a word character <==> [a-zA-Z0-9_]

\d – Matches digit character <==> [0-9]

\s – Matches whitespace character (space, tab, newline, etc.)


#### 13. What do means by \D, \W, and \S shorthand character classes signify in regular expressions?

Ans - The \D, \W, and \S shorthand character classes match a single character that is not a digit, word, or space characters.

\W – Matches any non-alphanumeric character <==> [^a-zA-Z0-9_]

\D – Matches any non-digit character <==> [^0-9]

\S – Matches any non-whitespace character
